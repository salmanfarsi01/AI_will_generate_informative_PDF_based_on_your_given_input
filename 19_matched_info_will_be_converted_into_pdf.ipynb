```python
import pandas as pd
from faker import Faker
import random
from datetime import date

fake = Faker()

# Predefined options
genders = ["Man", "Woman", "Any"]
locations = ["Stockholm", "Gothenburg", "Malmö", "Uppsala", "Lund"]
fields = ["Economics", "Technology", "Sports", "Art", "Science", "Medicine"]
funding_types = ["Scholarship", "Grant", "Sponsorship"]
renewable_options = ["Yes", "No"]

data = []

for _ in range(1000):
    row = {
        "Foundation Name": fake.company() + " Foundation",
        "Target Gender": random.choice(genders),
        "Target Age": f"{random.randint(18, 25)}-{random.randint(26, 40)}",
        "Location": random.choice(locations),
        "Field of Study": random.choice(fields),
        "Amount": f"{random.randint(5000, 20000)} SEK",
        "Application Deadline": fake.date_between(
            start_date=date(2025, 6, 1),
            end_date=date(2025, 12, 31)
        ).strftime("%Y-%m-%d"),  # convert to string!
        "Requirements": fake.sentence(nb_words=5),
        "Website": fake.url(),
        "Contact Email": fake.email(),
        "Max Applicants": random.randint(10, 500),
        "Funding Type": random.choice(funding_types),
        "Renewable": random.choice(renewable_options),
        "Notes": fake.sentence(nb_words=7)
    }
    data.append(row)

df = pd.DataFrame(data)
df.to_excel("dummy_foundations_50.xlsx", index=False)

print("✅ Dummy dataset with 50 rows saved as 'dummy_foundations_50.xlsx'")
```

```python
import pandas as pd
from fpdf import FPDF
import os
from google.generativeai import GenerativeModel

# Use the API key from env
api_key = os.getenv("GEMINI_API_KEY")

# Init Gemini
import google.generativeai as genai

genai.configure(api_key=api_key)  # ✅ set key ONCE

model = genai.GenerativeModel("gemini-2.5-flash")  # ✅ NO api_key here!


# Load Excel from /content/
df = pd.read_excel("dummy.xlsx")

# Get user input
target_gender = input("Gender (Man/Woman/Any): ")
location = input("Location (Stockholm, Gothenburg, Malmö, etc.): ")
field_of_study = input("Field of Study (Economics, Technology, etc.): ")

# Gemini prompt
prompt = f"""
Suggest keywords for:
Gender: {target_gender}
Location: {location}
Field: {field_of_study}
"""
response = model.generate_content(prompt)
print("Gemini suggestion:", response.text)

# Filter
filtered_df = df[
    (df["Target Gender"].str.lower() == target_gender.lower()) &
    (df["Location"].str.lower() == location.lower()) &
    (df["Field of Study"].str.lower() == field_of_study.lower())
]

print(f"Found {len(filtered_df)} matches!")

# PDF
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)
pdf.cell(200, 10, txt="Recommended Foundations", ln=True, align="C")

for idx, row in filtered_df.iterrows():
    pdf.ln(5)
    pdf.multi_cell(0, 10, txt=f"""
Foundation: {row['Foundation Name']}
Amount: {row['Amount']}
Deadline: {row['Application Deadline']}
Website: {row['Website']}
Contact: {row['Contact Email']}
Requirements: {row['Requirements']}
    """)

output_filename = "foundation_recommendations.pdf"
pdf.output(output_filename)
print(f"PDF saved: {output_filename}")

# Download link for Colab
from google.colab import files
files.download(output_filename)
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
from fpdf import FPDF
import google.generativeai as genai
import os

# 🔐 Gemini API
genai.configure(api_key="add your api")  # Replace with your real key
model = genai.GenerativeModel("gemini-2.5-flash")

# 📥 Load admin-uploaded Excel
df = pd.read_excel("/content/dummy.xlsx")  # Already uploaded in Colab

# 🎨 UI Layout
st.set_page_config(page_title="Foundation Finder", layout="centered")

st.markdown(
    "<h1 style='color: purple;'>🎓 Find your organization </h1>"
    "<p style='color: #666;'>Get personalized foundation recommendations based on your background.</p>",
    unsafe_allow_html=True
)

# 📋 User Form
gender = st.selectbox("Select Gender", ["Any", "Man", "Woman"])
location = st.text_input("Enter Location (e.g., Stockholm)")
field = st.text_input("Enter Field of Study (e.g., Technology)")

if st.button("🔍 Find Scholarships"):
    with st.spinner("🔮 Thinking..."):

        prompt = f"Suggest keywords for:\nGender: {gender}\nLocation: {location}\nField: {field}"
        response = model.generate_content(prompt)
        st.markdown(f"**🔮 AI BOT Suggestion:**\n\n{response.text}")

        filtered = df.copy()
        if gender != "Any":
            filtered = filtered[filtered["Target Gender"].str.lower() == gender.lower()]
        if location:
            filtered = filtered[filtered["Location"].str.lower() == location.lower()]
        if field:
            filtered = filtered[filtered["Field of Study"].str.lower() == field.lower()]

        if filtered.empty:
            st.warning("❌ No matching foundations found.")
        else:
            st.success(f"✅ Found {len(filtered)} matches!")
            st.dataframe(filtered)

            # 📄 Generate PDF
            pdf = FPDF()
            pdf.add_page()
            pdf.set_font("Arial", size=12)
            pdf.set_text_color(76, 0, 153)
            pdf.cell(200, 10, txt="Recommended Foundations", ln=True, align="C")
            pdf.set_text_color(0, 0, 0)

            for idx, row in filtered.iterrows():
                pdf.ln(5)
                pdf.multi_cell(0, 10, txt=f"""
Foundation: {row['Foundation Name']}
Amount: {row['Amount']}
Deadline: {row['Application Deadline']}
Website: {row['Website']}
Contact: {row['Contact Email']}
Requirements: {row['Requirements']}
                """)

            output_path = "foundation_recommendations.pdf"
            pdf.output(output_path)

            with open(output_path, "rb") as f:
                st.download_button("📥 Download PDF", f, file_name=output_path)

st.markdown("---")
st.markdown("Made with ❤️ using Streamlit + Gemini AI", unsafe_allow_html=True)
```
